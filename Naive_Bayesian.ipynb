```python
import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    'Age': ['Youth','Youth', 'Middle', 'Senior', 'Senior', 'Middle', 'Senior', 'Youth', 'Youth', 'Senior', 'Youth','Middle', 'Middle', 'Senior'],
    'Income': ['High', 'High', 'High', 'Medium', 'Low', 'Low', 'Low', 'Medium','LoW','Medium', 'Medium', 'Medium', 'High', 'Medium'],
    'Student': ['No', 'No', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'Yes','Yes','No', 'Yes', 'No'],
    'CreditRating': ['Fair', 'Excellent', 'Fair', 'Fair','Fair', 'Excellent', 'Excellent','Fair','Fair', 'Fair', 'Excellent', 'Excellent', 'Fair', 'Excellent'],
    'BuysComputer': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes','Yes', 'Yes', 'Yes', 'No']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
le = {}  # create empty dictionary
df_enc = df.copy()  # here we just cop our original data
for col in df.columns:  # it convert into the text using loop through of the c
    le[col] = LabelEncoder()  # now create new label for current data
    df_enc[col] = le[col].fit_transform(df[col])  # its learn unique value and
```

```python
df_enc
```

```python
x = df_enc[['Age', 'Income', 'Student', 'CreditRating']]
y = df_enc[['BuysComputer']]
```

```python
x
```

```python
model = CategoricalNB() #apply model
model.fit(x,y) # here fit your values
```

```python
test = [['Youth', 'Medium', 'Yes', 'Fair']] # testing data
```

```python
test
```

```python
print("\n------------------- Prediction -------------------\n")
for t in test:
    row = [[
        le['Age'].transform([t[0]])[0],
        le['Income'].transform([t[1]])[0],
        le['Student'].transform([t[2]])[0],
        le['CreditRating'].transform([t[3]])[0]
    ]]

    pred = le['BuysComputer'].inverse_transform(model.predict(row))[0]
    prob = model.predict_proba(row)[0]

    print(f"Input : Age={t[0]}, Income={t[1]}, Student={t[2]}, CreditRating={t[3]}")
    print(f"Result: Buys Computer -> {pred}")
    print(f"P(No)  = {prob[0]:.4f}")
    print(f"P(Yes) = {prob[1]:.4f}")
```
